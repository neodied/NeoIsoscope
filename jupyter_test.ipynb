```python
import requests
import json
from shtuff import POI

HERE_API_URL = r"https://isoline.route.cit.api.here.com/routing/7.2/calculateisoline.json"
auth = json.load(open("auth.json"))
```

```python
map_center_latlong_lst = [42.592416, -83.598935]
places = [POI(-83.598935, 42.592416, 15, alias="Home1"),
          POI(-83.444141, 42.413992, 25, alias="Home2"),
          POI(-83.674397, 42.571592, 15, alias="Work"),
          POI(-83.740355, 42.291279, 30, alias="A2"),
          ]

multilayer_map = folium.Map(location=map_center_latlong_lst,
                            zoom_start=10)

for place in places:
    payload = {'mode': 'fastest;car;traffic:disabled',
               'start': place.latlong_str,
               'rangetype': 'time',
               'range': place.timeradius_sec,
               'app_id': auth["HERE_API_App_ID"],
               'app_code': auth["HERE_API_App_Code"]}
    data = requests.get(HERE_API_URL, params=payload).json()
    place.poly = data["response"]["isoline"][0]["component"][0]["shape"]
    place.poly_geojson = json.dumps({
        "type": "Polygon",
        "coordinates": [
            [[x.split(',')[1], x.split(',')[0]] for x in place.poly]
        ]
    })
    folium.GeoJson(place.poly_geojson, name=place.alias).add_to(multilayer_map)
```

```python
folium.LayerControl().add_to(multilayer_map)
multilayer_map
```
