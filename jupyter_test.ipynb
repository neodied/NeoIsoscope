```python
import requests
import json
from shtuff import POI

HERE_API_URL = r"https://isoline.route.cit.api.here.com/routing/7.2/calculateisoline.json"
auth = json.load(open("auth.json"))

layers = [POI(-83.598935, 42.592416, 15, alias="Home"),
          POI(-83.444141, 42.413992, 15, alias="Lindsey")]
```

```python
payload = {'mode': 'fastest;car;traffic:disabled',
           'start': POI_home.latlong_str,
           'rangetype': 'time',
           'range': POI_home.timeradius_sec,
           'app_id': auth["HERE_API_App_ID"],
           'app_code': auth["HERE_API_App_Code"]}
r = requests.get(HERE_API_URL, params=payload)
data = r.json()
poly_home = data["response"]["isoline"][0]["component"][0]["shape"]
geo_json_poly_home = json.dumps({
    "type": "Polygon",
    "coordinates": [
        [[x.split(',')[1], x.split(',')[0]] for x in poly_home]
    ]
})
```

```python
home_15min_map = folium.Map(location=POI_home)
folium.GeoJson(geo_json_poly_home, name='geojson').add_to(home_15min_map)
folium.LayerControl().add_to(home_15min_map)
home_15min_map
```
